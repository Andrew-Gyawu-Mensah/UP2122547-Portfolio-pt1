```python
#Sets the variable to the input of the user
PrimeNum = int(input("Enter a number to be checked if it is prime or not: "))

#Checks if the input number is larger than 1, if it isn't it returns not prime as 1 and below aren't prime numbers
if PrimeNum > 1:
    
    
    #Checks if every whole number up to the input number is a factor by checking if the remainder of the modulus is 0, if it is it is not a prime number
    for i in range(2, PrimeNum):
        if (PrimeNum % i) == 0:
            print(PrimeNum, "is not a prime number")
            break
    else:
        print(PrimeNum, "is a prime number")
else:
    print(PrimeNum, "is not a prime number")
```

```python

#Sets the variable to the input of the user
nterms = int(input("How many terms? "))

#Sets two terms that are iteratively going to be added to the first two of the fibonacci sequence
n1, n2 = 0, 1

#Sets the count to 0
count = 0


#Checks if the number of terms is 0 or less, if so it returns an error and asks for input again until it works
while nterms <= 0:
   print("Please enter a positive integer")
   nterms = int(input("How many terms? "))

#If loop for just the first term of the fibonacci sequence 
if nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)

#Prints the current value of the sequence, sets the current values to the next value and makes the next values equal to the two values added
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1, end="   ")
       
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
#Takes the input from the user and sets it to a string
inputText = str(input("Enter a string you want to reverse: "))

#Function for reversing the text
def reverseText(inputText):
    
    #It starts from the end and goes towards the first taking each element. And therefore reverses the output
    return inputText[::-1]

print(reverseText(inputText))
```

```python
#Function to find the minimum and maximum values in a list of numbers
def find_min_max(numbers):
    
    #Returns a tuple with the min and max values
    return min(numbers), max(numbers)

#Sets users input to variable
user_input = input("Enter a list of numbers separated by spaces: ")

#Converts the user input string into a list of floating point numbers
numbers_list = [float(num) for num in user_input.split()]

#Calls on the function to get the minimum and maximum values
min_value, max_value = find_min_max(numbers_list)

#Print results
print("Min value is:", min_value)
print("Max value is:", max_value)
```

```python
import math

def series_sum(x, N):
    """
    Calculates the sum of the series:
      x(5^(21))/(1+2!) + x(5^(22))/(2+3!) + ... up to N terms.

    For the nth term:
      term = x * 5^(2n) / (n + (n+1)!)

    Parameters:
      x : number to be multiplied with each term.
      N : number of terms in the series.

    Returns:
      The sum of the series.
    """
    total = 0
    for n in range(1, N + 1):
        term = x (5 ** (2 * n)) / (n + math.factorial(n + 1))
        total += term
    return total

#Run the function with user inputs
x = float(input("Enter the value for x: "))
N = int(input("Enter the number of terms (N): "))

result = series_sum(x, N)
print("The sum of the series is:", result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
import urllib.request

#Load the image from the URL
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/68/Orange_tabby_cat_sitting_on_fallen_leaves-Hisashi-01A.jpg/1024px-Orange_tabby_cat_sitting_on_fallen_leaves-Hisashi-01A.jpg"
image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_data)
np_image = np.array(image)

#Create horizontal and vertical mirror reflections
horizontal_mirror = np.fliplr(np_image)
vertical_mirror = np.flipud(np_image)

#Display the original and mirrored images side by side
plt.figure(figsize=(18, 6))

#Original image
plt.subplot(1, 3, 1)
plt.imshow(np_image)
plt.title("Original Image")
plt.axis("off")

#Horizontal mirror image
plt.subplot(1, 3, 2)
plt.imshow(horizontal_mirror)
plt.title("Horizontal Mirror")
plt.axis("off")

#Vertical mirror image
plt.subplot(1, 3, 3)
plt.imshow(vertical_mirror)
plt.title("Vertical Mirror")
plt.axis("off")

plt.show()
```

```python
import numpy as np

# Step 1: Create a NumPy array with the given temperature data
temperatures_celsius = np.array([
    22, 24, 19, 21, 23, 25, 26, 27, 24, 23, 20, 22, 21, 19, 18, 
    22, 23, 24, 26, 28, 30, 27, 25, 26, 24, 22, 21, 20, 19, 23
])

# Step 2: Find min, max, and average temperature
min_temp = np.min(temperatures_celsius)
max_temp = np.max(temperatures_celsius)
avg_temp = np.mean(temperatures_celsius)

print(f"Minimum Temperature: {min_temp}C")
print(f"Maximum Temperature: {max_temp}C")
print(f"Average Temperature: {avg_temp:.2f}C")

# Step 3: Convert temperatures from Celsius to Fahrenheit
temperatures_fahrenheit = (temperatures_celsius * 9/5) + 32
print(f"Temperatures in Fahrenheit: {temperatures_fahrenheit}")

# Step 4: Find all days where the temperature was above 25C
above_25 = temperatures_celsius[temperatures_celsius > 25]
print(f"Days with temperature above 25C: {above_25}")

# Step 5: Count how many days had temperatures below 20C
below_20_count = np.sum(temperatures_celsius < 20)
print(f"Number of days with temperature below 20C: {below_20_count}")

# Step 6: Sort the temperature data in ascending order
sorted_temperatures = np.sort(temperatures_celsius)
print(f"Sorted Temperatures: {sorted_temperatures}")

# Step 7: Reshape the 1D array into a 5x6 2D array
reshaped_temperatures = temperatures_celsius[:30].reshape(5, 6)  # Taking first 30 elements to fit 5x6
print("Reshaped Temperature Data (5 weeks of 6 days each):")
print(reshaped_temperatures)
```
